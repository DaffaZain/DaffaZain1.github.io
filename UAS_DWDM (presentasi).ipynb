```python
import pandas as pd
import numpy as np 
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import mlxtend as ml
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_excel('transaksi_belanja_supermarket.xlsx')
data
```

```python
data = list(data["Items"].apply(lambda x:x.split(', ')))
data
```

```python
te = TransactionEncoder()
data_te = te.fit(data).transform(data)
data = pd.DataFrame(data_te,columns=te.columns_)
data
```

```python
data = data.astype(int)
data
#bilangan biner
```

```python
data.to_excel('2010512030_Daffa Andika Zain_UAS.xlsx', index=1)
```

```python
data.iloc[:,:].sum()
```

```python
frequent_itemsets = apriori(data, min_support=0.5, use_colnames=True)
frequent_itemsets.head()
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
rules.shape
```
